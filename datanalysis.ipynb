```python
import pandas as pd
myfiles= pd.read_csv("C:\\Users\\Manasa\\OneDrive\\Desktop\\myfiles\\data2.csv")
print(myfiles.head())
print(myfiles.info())   
print(myfiles.describe()) 
product_sales = myfiles.groupby("Product")["Sales"].sum()
print(product_sales)
region_sales =myfiles.groupby("Region")["Sales"].sum()
print(region_sales)
import matplotlib.pyplot as plt
product_sales.plot(kind="bar", title="Sales by Product")
plt.show()
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Sales by Region")
plt.ylabel("")  
plt.show()
```
